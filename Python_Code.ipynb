```python
import requests

api_key = "10b0b3741256c3d840ad164ba3cba2c8"

def get_weather(city):
    url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"
    data = requests.get(url).json()
    return {
        'City': city,
        'Temp': data['main']['temp'],
        'Humidity': data['main']['humidity'],
        'Weather': data['weather'][0]['description']
    }
```

```python
top_cities = ['Kolkata', 'Mumbai', 'Delhi', 'Bengaluru', 'Chennai']
weather_data = [get_weather(city) for city in top_cities]
```

```python
import matplotlib.pyplot as plt

# Example: Temperature
cities = [d['City'] for d in weather_data]
temps = [d['Temp'] for d in weather_data]

plt.bar(cities, temps, color='orange')
plt.title("Temperature Comparison Across Cities")
plt.ylabel("Temperature (°C)")
plt.xticks(rotation=45)
plt.show()
```
